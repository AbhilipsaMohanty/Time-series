# 2: Model Building - ARIMA

### 1.	Parameter Selection for ARIMA

#### ACF and PACF plots 

```python
plt.figure(figsize=(12, 8))
plt.subplot(121)
plot_acf(data, ax=plt.gca(), lags=20)
plt.title('ACF Plot')
plt.subplot(122)
plot_pacf(data, ax=plt.gca(), lags=30)
plt.title('PACF Plot')

plt.show()
```

### 2.	Model Fitting

```python
model= ARIMA(train,order=(1,1,1))
model_fit=model.fit()
```

```python
print(model_fit.summary())
```

### 3.	Diagnostics

```python
residuals = model_fit.resid
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(residuals)
plt.title('Residuals')

plt.subplot(1, 2, 2)
plot_acf(residuals, lags=30, ax=plt.gca())
plt.title('ACF of Residuals')

plt.tight_layout()
plt.show()
```

### 4.	Forecasting

```python
test
```

```python
pred_test=model_fit.forecast(steps=len(test))
pred_test
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train, label='Train Data')
plt.plot(test, label='Test Data', color='green')
plt.plot(pred_test, label='Forecast', color='red', linestyle='--')

# Adding labels and legend
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Forecast')
plt.legend()
plt.show()
```

# 4: Evaluation and Comparison

####  1.Compute Error Metrics

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    nonzero_indices = y_true != 0
    return np.mean(np.abs((y_true- y_pred) / y_true)) * 100

def compute_metrics(actual, forecast):
    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    mape = mean_absolute_percentage_error(actual, forecast)
    return mae, rmse, mape
```

```python
mae_simple, rmse_simple, mape_simple = compute_metrics(test, forecast)
mae_holt, rmse_holt, mape_holt = compute_metrics(test, forecast1)
mae_winter, rmse_winter, mape_winter = compute_metrics(test, forecast2)
mae_arima, rmse_arima, mape_arima = compute_metrics(test,pred_test)
```

```python
print(f"Simple Exponential Smoothing - MAE: {mae_simple:.2f}, RMSE: {rmse_simple:.2f}, MAPE: {mape_simple:.2f}%")
print(f"Holt's Linear Trend Model - MAE: {mae_holt:.2f}, RMSE: {rmse_holt:.2f}, MAPE: {mape_holt:.2f}%")
print(f"Holt-Winters Model - MAE: {mae_winter:.2f}, RMSE: {rmse_winter:.2f}, MAPE: {mape_winter:.2f}%")
print(f"ARIMA Model - MAE: {mae_arima:.2f}, RMSE: {rmse_arima:.2f}, MAPE: {mape_arima:.2f}%")
```

#### 2.	Model Comparison

The ARIMA Model shows the best overall performance with the lowest MAPE (9.06%) and equal lowest RMSE (0.07) and MAE (0.07), indicating high accuracy. The Holt-Winters Model is also strong, particularly in handling seasonality, with comparable MAE (0.07) and RMSE (0.07) and a slightly higher MAPE (9.36%). Simple Exponential Smoothing performs well but slightly less effectively, particularly in RMSE (0.08) and MAPE (9.55%). Holt's Linear Trend Model lags significantly behind, with higher MAE (0.10), RMSE (0.14), and MAPE (15.55%), indicating poorer trend handling.

#### 3.	Conclusion

The ARIMA Model stands out as the most accurate, with the lowest MAPE (9.06%) and matching the best RMSE (0.07) and MAE (0.07), making it ideal for precise forecasting. The Holt-Winters Model also performs well, particularly in seasonal data, with comparable metrics. Simple Exponential Smoothing is effective but slightly less accurate, particularly in RMSE (0.08) and MAPE (9.55%). Holt's Linear Trend Model shows the weakest performance, with higher errors, making it less suitable for accurate forecasting.
